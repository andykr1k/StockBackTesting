```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
start = datetime.datetime(2020,1,1)
end = datetime.datetime(2022,10,5)
```

```python
googl = web.DataReader("GOOGL", "yahoo", start, end)
aapl = web.DataReader("AAPL", "yahoo", start, end)
tsla = web.DataReader("TSLA", "yahoo", start, end)
msft = web.DataReader("MSFT", "yahoo", start, end)
spy = web.DataReader("SPY", "yahoo", start, end)
```

```python
spy.reset_index()
aapl.reset_index()
msft.reset_index()
googl.reset_index()
tsla.reset_index()
```

```python
spy.head()
```

```python
spy["Close"].plot(label='Spy', figsize=(12,6))
tsla["Close"].plot(label='Tesla')
aapl["Close"].plot(label='Apple')
googl["Close"].plot(label='Google')
msft["Close"].plot(label='Microsoft')
plt.legend()
```

```python
spy["Volume"].plot(label='Spy', figsize=(12,6))
tsla["Volume"].plot(label='Tesla')
aapl["Volume"].plot(label='Apple')
googl["Volume"].plot(label='Google')
msft["Volume"].plot(label='Microsoft')
plt.legend()
```

```python
spy['Total $ Traded'] = spy['Close'] * spy['Volume']
aapl['Total $ Traded'] = aapl['Close'] * aapl['Volume']
msft['Total $ Traded'] = msft['Close'] * msft['Volume']
googl['Total $ Traded'] = googl['Close'] * googl['Volume']
tsla['Total $ Traded'] = tsla['Close'] * tsla['Volume']
spy.head()
```

```python
spy["Total $ Traded"].plot(label='Spy', figsize=(12,6))
tsla["Total $ Traded"].plot(label='Tesla')
aapl["Total $ Traded"].plot(label='Apple')
googl["Total $ Traded"].plot(label='Google')
msft["Total $ Traded"].plot(label='Microsoft')
plt.legend()
```

```python
spy['MA5'] = spy['Close'].rolling(5).mean()
aapl['MA5'] = aapl['Close'].rolling(5).mean()
msft['MA5'] = msft['Close'].rolling(5).mean()
googl['MA5'] = googl['Close'].rolling(5).mean()
tsla['MA5'] = tsla['Close'].rolling(5).mean()
spy.tail()
```

```python
spy['MA9'] = spy['Close'].rolling(9).mean()
aapl['MA9'] = aapl['Close'].rolling(9).mean()
msft['MA9'] = msft['Close'].rolling(9).mean()
googl['MA9'] = googl['Close'].rolling(9).mean()
tsla['MA9'] = tsla['Close'].rolling(9).mean()
spy.tail()
```

```python
spy['MA50'] = spy['Close'].rolling(50).mean()
aapl['MA50'] = aapl['Close'].rolling(50).mean()
msft['MA50'] = msft['Close'].rolling(50).mean()
googl['MA50'] = googl['Close'].rolling(50).mean()
tsla['MA50'] = tsla['Close'].rolling(50).mean()
spy.tail()
```

```python
spy['MA200'] = spy['Close'].rolling(200).mean()
aapl['MA200'] = aapl['Close'].rolling(200).mean()
msft['MA200'] = msft['Close'].rolling(200).mean()
googl['MA200'] = googl['Close'].rolling(200).mean()
tsla['MA200'] = tsla['Close'].rolling(200).mean()
spy.tail()
```

```python
spy['Close'].plot(label="Spy Closing Prices", figsize=(16,8))
spy['MA5'].plot(label="MA5")
spy['MA9'].plot(label="MA9")
spy['MA50'].plot(label="MA50")
spy['MA200'].plot(label="MA200")
plt.legend()
```

```python
aapl['Close'].plot(label="Apple Closing Prices", figsize=(16,8))
aapl['MA5'].plot(label="MA5")
aapl['MA9'].plot(label="MA9")
aapl['MA50'].plot(label="MA50")
aapl['MA200'].plot(label="MA200")
plt.legend()
```

```python
spy['Close'].plot(label="Spy Closing Prices", figsize=(16,8))
spy['MA200'].plot(label="Spy MA200")
spy['MA50'].plot(label="Spy MA50")
spy['MA9'].plot(label="Spy MA9")
googl['MA200'].plot(label="Google MA200")
googl['MA50'].plot(label="Google MA50")
googl['MA9'].plot(label="Google MA9")
tsla['MA200'].plot(label="Tesla MA200")
tsla['MA50'].plot(label="Tesla MA50")
tsla['MA9'].plot(label="Tesla MA9")
aapl['MA200'].plot(label="Apple MA200")
aapl['MA50'].plot(label="Apple MA50")
aapl['MA9'].plot(label="Apple MA9")
msft['MA200'].plot(label="Microsoft MA200")
msft['MA50'].plot(label="Microsoft MA50")
msft['MA9'].plot(label="Microsoft MA9")
plt.legend()
```

```python
spy['RVOL'] = spy['Volume'] / spy['MA50']
aapl['RVOL'] = aapl['Volume'] / aapl['MA50']
msft['RVOL'] = msft['Volume'] / msft['MA50']
googl['RVOL'] = googl['Volume'] / googl['MA50']
tsla['RVOL'] = tsla['Volume'] / tsla['MA50']
spy.tail()
```

```python
spy['Close'].plot(label="Spy Closing Prices", figsize=(16,8))
spy['MA200'].plot(label="Spy MA200")
spy['MA50'].plot(label="Spy MA50")
plt.legend()
plt.figure()
spy['RVOL'].plot(label="Spy RVOL", figsize=(16,4))
plt.legend()
```

```python
spy['% Change'] = (spy['Close']/spy['Open']) / spy['Open'] * 100
spy['Range'] = (spy['High']-spy['Low'])
spy.tail()
```

```python
for items in spy:
    if (spy.iloc[items]['Low'] > spy.iloc[items - 1]['Low'] and spy.iloc[items]['High'] < spy.iloc[items - 1]['High']):
        spy['Strat'] = "Inside"
    elif (spy.iloc[items]['Low'] < spy.iloc[items - 1]['Low'] and spy.iloc[items]['High'] > spy.iloc[items - 1]['High']):
        spy['Strat'] = "Outside"
    elif (spy.iloc[items]['Low'] > spy.iloc[items - 1]['Low'] and spy.iloc[items]['High'] > spy.iloc[items - 1]['High']):
        spy['Strat'] = "Upside"
    elif (spy.iloc[items]['Low'] < spy.iloc[items - 1]['Low'] and spy.iloc[items]['High'] < spy.iloc[items - 1]['High']):
        spy['Strat'] = "Downside"
```
